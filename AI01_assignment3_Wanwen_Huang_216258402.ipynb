### Description

```CarRacing-v0``` is one of the gym's easiest tasks for learning continuous control task to learn from pixels. *Continous* control means that instead of discret actions (e.g. Pong's up/down), here actions continous with a scalar magnitude can be taken. There are three different actions for the car: **steer** which can take values from -1. to 1., **gas** (0. -- 1.) & **brake** (0. -- 1.).
The agent (car) observes the environment from above (not very naturalisitc), and each observation consists of a 96x96 pixels image. Reward is -0.1 every step and +1000/N for every track tile visited, where N is the total number of tiles in track. For example, if you have finished in 732 step, your reward is 1000 - 0.1*732 = 926.8 points. Some indicators shown at the bottom of the observation and the state RGB buffer. From left to right: true speed, four ABS sensors, steering wheel position, gyroscope.

## Discounted reward function

````python
def discount_and_normalize_rewards(episode_rewards, gamma=0.95):
    """
    """

    discounted_episode_rewards = np.zeros(episode_rewards.shape)
    cumulative = 0.0
            
    for i in reversed(range(episode_rewards.shape[0])):
        
        cumulative = cumulative * gamma + episode_rewards[i]
        discounted_episode_rewards[i] = cumulative
    
    mean = discounted_episode_rewards.mean()
    std = discounted_episode_rewards.std()
    discounted_episode_rewards = (discounted_episode_rewards - mean) / std
    
    return discounted_episode_rewards
````

## State preprocessing
* Crop
* RGB -> gray-scale
* Stack 4 consecutive frames into one state

````python
def preprocess_state(previous_state, observation):
    """
    Argument
    --------    observation : An image with shape = (96, 96, 3) and dype = np.uint8
    previous_state : A 3d array with shape = (1, 84, 84, 4), 
                     where consecutive observations are held in the 3rd dimension.
    
    Return
    ------
    
    next_state : A 3d array with shape = (1, 84, 84, 4) & dtype=np.float64,
                 where consecutive observations are held in the 3rd dimension. and dtype = np.float64
                 
    """   
    
    img = observation
    img = img[1:85,6:90]
    tmp = cv2.resize(src=img, dsize=(84,84))
    tmp = cv2.cvtColor(tmp, cv2.COLOR_BGR2GRAY)
    # change the dtype
    tmp = cv2.normalize(tmp, tmp, alpha=0.0, beta=1.0, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_64F)
    # expand one dimension (84, 84, 1)
    tmp = np.expand_dims(tmp, len(tmp.shape))
    #> tmp.shape = (84, 84, 1)
    # squeeze the previous state arrays shape = (1,84,84,4) into 3d
    previous_state_sq = np.squeeze(previous_state, axis=0)
    # previous_state_sq shape = (84,84,4)
    # stack new obs
    previous_state_sq = np.dstack((previous_state_sq, tmp))
    # new shape -> (84, 84 ,5)
    # replace the oldest observation
    previous_state_sq = previous_state_sq[:,:,1:]
    #add a new dimension
    next_state = np.expand_dims(previous_state_sq, axis=0)

    return next_state
````

## Settings

### Agent
The agent can take 3 actions, and each action is continous.
* streer (-1. to 1.)
* gas (0. to 1.)
* brake (0. to 1.)

You can check how many actions can be taken with ```env.action_space.shape```.
The range for each action can be viewed with ```env.action_space.low```  and ```env.action_space.high```.

However, brake and gas are antagonistic, that is, there is no point in breaking while driving at full gas. So, in order to simplify the model a little we can combine gas and brake into on action which we can call **gake**. Gake can take values from -1. to 1., where negative values activates the brake, and positive the gas. The policy gradient model is trained with actions in a two column format (```steer```, ```gake```), whereas the action input to ```env.step()``` is in a three-column format (```steer```, ```gas```, ```brake```).


### State
The state has ```shape = (84, 84, 4)```. The first two dimensions holds an observaion (84 x 84 pixels), and the 3rd dim holds a stack of 4 consecutive observations in order to give the network information about changes between observations.

````python
# Number of actions in model format, i.e. steer & gake
n_actions_model_fmt = 2

# State size
state_shape = [84, 84, 4]
# Model input_shape (n_samples, n_rows, n_columns, n_channels)
input_shape = (1, *state_shape)
# Initialize the state with zeros
state = np.zeros(shape=input_shape)

# Hyperparameters
batch_sz = 4

n_episodes = 1000

learning_rate = 0.001

# Location of game videos
video_dir = 'video'
# Location and base name of saved models
# If you want to resume training of a pre-existing model,
# then you will need to set this path to the name of the 
# model of interest. 
# E.g. for the model saved at the 100th episode:
#      model_path = "%s/pg_model-100" % model_dir
model_dir = './model'
model_path = "%s/pg_model" % model_dir

# resume training of a pre-existing model
resume = False # True | False
````

## Build model

````python
class PGModel:
    
    def __init__(self, env, state_shape, input_shape, n_actions, learning_rate, name='PGModel'):
        """
        Policy Gradient Model
        """
                
        self.input_shape = input_shape
        self.state_shape = state_shape
        self.n_actions = n_actions
        self.learning_rate = learning_rate
        
        with tf.variable_scope(name):
            with tf.name_scope("inputs"):
                # We create the placeholders
                # *state_shape means that we take each elements of state_shape in tuple hence is like if we wrote
                # [None, 84, 84, 5]
                self.input_= tf.placeholder(tf.float32, [None, *state_shape], name="input_")
                self.actions = tf.placeholder(tf.float32, [None, n_actions], name="actions")
                self.discounted_rewards = tf.placeholder(tf.float32, [None, 1], 
                                                         name="discounted_rewards")
            
                
                # Add this placeholder for viewing this variable in tensorboard
                self.mean_reward_ = tf.placeholder(tf.float32, name="mean_reward")
                
            with tf.name_scope("conv1"):
                """
                First Convolutional layer followed by BatchNormalization
                """
                # Input shape is (84, 84, 5)
                self.conv1 = tf.layers.conv2d(inputs=self.input_,
                                              filters=32,
                                              kernel_size=[8,8],
                                              strides=[4,4],
                                              padding="VALID",
                                              kernel_initializer=tf.contrib.layers.xavier_initializer_conv2d(),
                                              name="conv1")

                self.conv1_batchnorm = tf.layers.batch_normalization(self.conv1,
                                                                     training=True,
                                                                     epsilon=1e-5,
                                                                     name='batchnorm1')

                self.conv1_out = tf.nn.elu(self.conv1_batchnorm, name="conv1_out")
                # shape (20, 20, 32)
            
            with tf.name_scope("conv2"):
                """
                Second Convolutional layer followed by BatchNormalization
                """
                self.conv2 = tf.layers.conv2d(inputs=self.conv1_out,
                                              filters=64,
                                              kernel_size=[4,4],
                                              strides=[2,2],
                                              padding="VALID",
                                              kernel_initializer=tf.contrib.layers.xavier_initializer_conv2d(),
                                              name="conv2")

                self.conv2_batchnorm = tf.layers.batch_normalization(self.conv2,
                                                                     training=True,
                                                                     epsilon=1e-5,
                                                                     name='batchnorm2')

                self.conv2_out = tf.nn.elu(self.conv2_batchnorm, name="conv2_out")
                # shape: (9, 9, 64)
            
            with tf.name_scope("conv3"):
                """
                Third Convolutional layer followed by BatchNormalization
                """
                self.conv3 = tf.layers.conv2d(inputs=self.conv2_out,
                                              filters=128,
                                              kernel_size=[4,4],
                                              strides=[2,2],
                                              padding="VALID",
                                              kernel_initializer=tf.contrib.layers.xavier_initializer_conv2d(),
                                              name="conv3")

                self.conv3_batchnorm = tf.layers.batch_normalization(self.conv3,
                                                                     training=True,
                                                                     epsilon=1e-5,
                                                                     name='batchnorm3')

                self.conv3_out = tf.nn.elu(self.conv3_batchnorm, name="conv3_out")
                # shape: (3, 3, 128)
            
            with tf.name_scope("flatten"):
                """
                Flatten the 3D conv3_out to prepare it for a fully connected layer
                """
                self.flatten = tf.layers.flatten(self.conv3_out)
                # shape: (1152)
            
            with tf.name_scope("fc"):
                """
                Final, fully connected layer
                """
                self.fc = tf.layers.dense(inputs=self.flatten,
                                          units=512,
                                          activation=tf.nn.elu,
                                          kernel_initializer=tf.contrib.layers.xavier_initializer(),
                                          name="fc")
            
            with tf.name_scope("logits"):
                """
                Output of the fully connected layer
                """
                self.logits = tf.layers.dense(inputs=self.fc, 
                                              kernel_initializer=tf.contrib.layers.xavier_initializer(),
                                              units=self.n_actions * 2, 
                                              activation=None)
            
            with tf.name_scope("softmax"):
                self.action_distribution = tf.nn.softmax(self.logits)
                

            with tf.name_scope("loss"):
                            
                ## Steering
                self.steer_m = self.logits[:, 0]  # mean
                self.steer_sd = self.logits[:, 1] # standard deviation
                self.steer_sd = tf.nn.softplus(self.steer_sd) + 1e-5
                # normal distribution from mean and standard deviation
                self.steer_normal_dist = tf.distributions.Normal(loc=self.steer_m, scale=self.steer_sd)                
                # sample the distribution
                self.steer_action = self.steer_normal_dist._sample_n(1)
                # clip so that the values stay within the range accepted by env.step()
                self.steer_action = tf.clip_by_value(self.steer_action, 
                                                     env.action_space.low[0], 
                                                     env.action_space.high[0])

                ## Combined gas and brake: gake
                self.gake_m = self.logits[:, 2]  # mean
                self.gake_sd = self.logits[:, 3] # standard deviation
                self.gake_sd = tf.nn.softplus(self.gake_sd) + 1e-5
                # normal distribution from mean and standard deviation
                self.gake_normal_dist = tf.distributions.Normal(loc=self.gake_m, scale=self.gake_sd)
                self.gake_action = self.gake_normal_dist._sample_n(1)   
                # range is -1 to 1, where negative values is brake and positive is gas
                self.gake_action = tf.clip_by_value(self.gake_action, 
                                                    -env.action_space.high[2], 
                                                    env.action_space.high[1])
                
                # Cross entropy loss to encourage exploration
                self.steer_loss = -self.steer_normal_dist.log_prob(self.steer_action) * self.actions[:, 0]
                self.steer_loss -= 1e-1 * self.steer_normal_dist.entropy()
                self.gake_loss = -self.gake_normal_dist.log_prob(self.gake_action) * self.actions[:, 1]
                self.gake_loss -= 1e-1 * self.gake_normal_dist.entropy()
                # Add cross entropy cost to encourage exploration
                self.loss = tf.reduce_mean((self.steer_loss + self.gake_loss) * self.discounted_rewards)
                
                
            
            with tf.name_scope("train"):
                self.optimizer = tf.train.AdamOptimizer(learning_rate=self.learning_rate)
                self.train_op = self.optimizer.minimize(self.loss, 
                                                        global_step=tf.train.get_global_step())
        

    
    def predict(self, state, sess=None):
        """
        """
        sess = sess or tf.get_default_session()

        # Get a probability distribution over the possible actions
        steer_action, gake_action = sess.run([self.steer_action, self.gake_action],
                                             feed_dict={self.input_: state.reshape(self.input_shape)})
        
        # model-format: 2-columns, [steer, gake]
        model_action = np.hstack([steer_action, gake_action])
        
        # Agent/car format: 3-columns, [steer, gas, brake]
        # Split up gake_action into gas and brake actions
        b = gake_action < 0
        brake_action = - gake_action.copy()
        brake_action[np.logical_not(b)] = 0.
        gas_action = gake_action.copy()
        gas_action[b] = 0.
        
        # Combine them into a 3-column array
        action = np.hstack([steer_action, gas_action, brake_action])
        
        return action.ravel(), model_action.ravel()
````

## Model and environment initialization

````python
# Reset the graph
tf.reset_default_graph()

# Initialize environment
env = gym.make('CarRacing-v0')
env = wrappers.Monitor(env, video_dir, force = True) # no previous video

# Instantiate the PGModel
model = PGModel(env, state_shape, input_shape, n_actions_model_fmt, learning_rate)

# Add ops to save and restore all the variables.
saver = tf.train.Saver()

# Store state, action & reward in a batch for later learning
episode_rewards, batch_losses = [], []
batch_store = {'state': [], 'action': [], 'discounted_reward': []}
````

## Play a game
Plays one episode (i.e. a game).

````python
def play_episode(env, state, model, gamma=0.95):
    """
    """
    
    env.reset()
    
    # Store state, action & reward in a batch for later learning
    store = {'state': [], 'action': [], 'reward': []}
    
    # Begin with no action just to get an observation
    observation, reward, done, info = env.step([0., 0., 0.])
    
    while not done: # while this episode is still running

        # Add the current observation [shape: (84, 84)] to the state [shape: (84, 84, 5)]
        state = preprocess_state(state, observation) 
        
        # Get action
        action, tf_action = model.predict(state)

        # Take action
        observation, reward, done, info = env.step(action)

        # ------------------------------            
        # Store state, action, & reward
        # ------------------------------
        # State        
        store['state'].append(state)
        # Action, in model format, i.e. two columns: steer & gas, 
        # where neg values for gas translates to break & positive translates to gas
        store['action'].append(tf_action)
        # Reward
        store['reward'].append(reward)
        
    # Calculate the discounted rewards once the episode is over
    store['discounted_reward'] = discount_and_normalize_rewards(np.vstack(store['reward']))
        
            
    return store 
````

## Update the weights

````python
def apply_gradients(batch):
    """
    Update the weights
    """
            
    loss_, _ = sess.run([model.loss, model.train_op], 
                        feed_dict={model.input_: np.vstack(batch['state']),
                                   model.actions: np.vstack(batch['action']),
                                   model.discounted_rewards: np.vstack(batch['discounted_reward'])})
    return loss_
````
